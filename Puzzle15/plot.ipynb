```python
import matplotlib.pyplot as plt
```

```python
INPUT = {
    'MD' : 'out1.txt',
    'ID' : 'out2.txt',
    'max(MD, ID)' : 'out3.txt',
}

VISITED = {}
SOL_LEN = {}
TIME = {}

for heuristic, file_name in INPUT.items():
    with open(file_name, 'r') as file:
        VISITED[heuristic] = []
        SOL_LEN[heuristic] = []
        TIME[heuristic] = []
        for line in file.readlines():
            rounds, path_len, time = line.split(',')
            VISITED[heuristic].append(int(rounds.strip()))
            SOL_LEN[heuristic].append(int(path_len.strip()))
            TIME[heuristic].append(int(time.strip()))
```

```python
for heuristic, data in VISITED.items():
    avg = sum(data) / len(data)

    plt.hist(data, bins=10)
    plt.axvline(avg, color='k', linestyle='dashed', linewidth=1)

    min_ylim, max_ylim = plt.ylim()
    plt.text(avg*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(avg))

    plt.title('Liczba odwiedzonych stanów dla heurystyki ' + heuristic)
    plt.figure(figsize=(10,8))
    plt.show()
```

```python
for heuristic, data in SOL_LEN.items():
    avg = sum(data) / len(data)

    plt.hist(data, bins=10)
    plt.axvline(avg, color='k', linestyle='dashed', linewidth=1)

    min_ylim, max_ylim = plt.ylim()
    plt.text(avg*1.01, max_ylim*0.9, 'Mean: {:.2f}'.format(avg))

    plt.title('Długość rowiązania dla heurystyki ' + heuristic)
    plt.figure(figsize=(10,8))
    plt.show()
```

```python
heuristics = VISITED.keys()
for heuristic in heuristics:
    rounds = VISITED[heuristic]
    sol_lens = SOL_LEN[heuristic]
    
    plt.title('Liczba odwiedzonych stanów vs\n długość rowiązania dla heurystyki ' + heuristic)
    plt.scatter(sol_lens, rounds, marker='x')
    plt.figure(figsize=(10,8))
    plt.show()

```

```python
for heuristic, data in TIME.items():
    avg = sum(data) / len(data)

    plt.hist(data, bins=10)
    plt.axvline(avg, color='k', linestyle='dashed', linewidth=1)

    min_ylim, max_ylim = plt.ylim()
    plt.text(avg*1.01, max_ylim*0.9, 'Mean: {:.2f}'.format(avg))

    plt.title('Czas rowiązania dla heurystyki ' + heuristic + ' [sekundy]')
    plt.figure(figsize=(10,8))
    plt.show()
```
